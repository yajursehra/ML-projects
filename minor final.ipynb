```python
fd = pd.DataFrame(zip(df.male, df.age, df.education, df.currentSmoker, df.cigsPerDay, df.BPMeds, df.prevalentStroke, df.prevalentHyp, df.diabetes, df.totChol, df.sysBP, df.diaBP, df.BMI, df.heartRate, df.glucose, df.TenYearCHD)) 
```

```python
fd.columns=['male','age','education','currentsmoker','cigsperday','bpmeds','prevalentstroke','prevalenthyp','diabetes','totchol','sysbp','diabp','bmi','heartrate','glucose','TenYearCHD']
```

```python
feature_cols = ['male','age','currentsmoker','cigsperday','bpmeds','prevalentstroke','prevalenthyp','diabetes','totchol','sysbp','diabp','bmi','heartrate','glucose']
```

```python
X = fd[feature_cols] 
y = fd.TenYearCHD
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```
